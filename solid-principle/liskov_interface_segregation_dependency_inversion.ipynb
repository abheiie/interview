## Liskov substitution 

     Child should have all the functionalities of parent and perform it as expected ðŸ˜ª no surprises

     Whenever we subtitute a parent class with child class, we should insure that we do not 
     alter the correctness of the program.

     We should be able to use instance of child class in place of instace of parent class.

     Any derive class should be able to implement the super class method availabel to it.

````python
"""
TODO SOLUTION 1 -
"""

from enum import Enum
from abc import ABC, abstractmethod

class BirdType(Enum):
    EAGLE = "EAGLE"
    PEACOCK = "PEACOCK"
    DOVE = "DOVE"
    SPARROW = "SPARROW"
    PARROT = "PARROT"
    PENGUIN = "PENGUIN"

class Size(Enum):
    SMALL = 'small'
    MEDIUM = 'medium'
    LARGE = 'large'

class Color(Enum):
    """
    Color of bird
    """
    RED = 'red'
    BLUE = 'blue'
    GREEN = 'green'


class Bird(ABC):
    def __init__(self, weight, colour, size, bird_type):
        self.weight = weight
        self.colour = colour
        self.size = size
        self.type = bird_type


class Flyable(ABC):

    @abstractmethod
    def fly(self):
        pass


class NonFlyable(ABC):
    pass


class Eagle(Bird, Flyable):
    def __init__(self, weight, colour, size, bird_type):
        super().__init__(weight, colour, size, bird_type)

    def fly(self):
        print("\nEagle is flying")


class Sparrow(Bird, Flyable):
    def __init__(self, weight, colour, size, bird_type):
        super().__init__(weight, colour, size, bird_type)

    def fly(self):
        print("\nSparrow is flying")


class Penguin(Bird, NonFlyable):
    def __init__(self, weight, colour, size, bird_type):
        super().__init__(weight, colour, size, bird_type)


"""
As now we have solved Substitution problem 
"""

eagle = Eagle(weight=10, colour=Color.BLUE, size=Size.LARGE, bird_type=BirdType.EAGLE)
sparrow = Sparrow(weight=10, colour=Color.RED, size=Size.SMALL, bird_type=BirdType.SPARROW)


flyable_birds = [eagle, sparrow]

for bird in flyable_birds:
    bird.fly()
````
