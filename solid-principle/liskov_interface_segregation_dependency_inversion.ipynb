````python
"""
TODO SOLUTION 1 -
"""

from enum import Enum
from abc import ABC, abstractmethod

class BirdType(Enum):
    EAGLE = "EAGLE"
    PEACOCK = "PEACOCK"
    DOVE = "DOVE"
    SPARROW = "SPARROW"
    PARROT = "PARROT"
    PENGUIN = "PENGUIN"

class Size(Enum):
    SMALL = 'small'
    MEDIUM = 'medium'
    LARGE = 'large'

class Color(Enum):
    """
    Color of bird
    """
    RED = 'red'
    BLUE = 'blue'
    GREEN = 'green'


class Bird(ABC):
    def __init__(self, weight, colour, size, bird_type):
        self.weight = weight
        self.colour = colour
        self.size = size
        self.type = bird_type


class Flyable(ABC):

    @abstractmethod
    def fly(self):
        pass


class NonFlyable(ABC):
    pass


class Eagle(Bird, Flyable):
    def __init__(self, weight, colour, size, bird_type):
        super().__init__(weight, colour, size, bird_type)

    def fly(self):
        print("\nEagle is flying")


class Sparrow(Bird, Flyable):
    def __init__(self, weight, colour, size, bird_type):
        super().__init__(weight, colour, size, bird_type)

    def fly(self):
        print("\nSparrow is flying")


class Penguin(Bird, NonFlyable):
    def __init__(self, weight, colour, size, bird_type):
        super().__init__(weight, colour, size, bird_type)


"""
As now we have solved Substitution problem 
"""

eagle = Eagle(weight=10, colour=Color.BLUE, size=Size.LARGE, bird_type=BirdType.EAGLE)
sparrow = Sparrow(weight=10, colour=Color.RED, size=Size.SMALL, bird_type=BirdType.SPARROW)


flyable_birds = [eagle, sparrow]

for bird in flyable_birds:
    bird.fly()
````

## Dependency Inversion Principle

-   High level module should not depend on low level module, they should depend on Abstraction

````python
"""
Lets create a Beak class(Abstract) having subclasses StrongBeak and WeakBeak, 
what problem we will face if we try to associate subclass of Beak (StrongBeak, WeakBeak) 
in our Bird class.
"""

from enum import Enum
from abc import ABC, abstractmethod

class BirdType(Enum):
    EAGLE = "EAGLE"
    PEACOCK = "PEACOCK"
    DOVE = "DOVE"
    SPARROW = "SPARROW"
    PARROT = "PARROT"
    PENGUIN = "PENGUIN"

class Size(Enum):
    SMALL = 'small'
    MEDIUM = 'medium'
    LARGE = 'large'

class Color(Enum):
    """
    Color of bird
    """
    RED = 'red'
    BLUE = 'blue'
    GREEN = 'green'

class Beak(ABC):
    def __init__(self, strength) -> None:
        self.strength = strength

class StrongBeak(Beak):
    def __init__(self, strength) -> None:
        super().__init__(strength)

class WeakBeak(Beak):
    def __init__(self, strength) -> None:
        super().__init__(strength)


class Bird(ABC):
    def __init__(self, weight, colour, size, bird_type, beak):
        self.weight = weight
        self.colour = colour
        self.size = size
        self.type = bird_type

        # Associate subclasses of Beak in Bird class
        self.beak: StrongBeak = beak


class Flyable(ABC):

    @abstractmethod
    def fly(self):
        pass


class NonFlyable(ABC):
    pass


class Eagle(Bird, Flyable):
    def __init__(self, weight, colour, size, bird_type, beak):
        super().__init__(weight, colour, size, bird_type, beak)

    def fly(self):
        print("\nEagle is flying")


class Sparrow(Bird, Flyable):
    def __init__(self, weight, colour, size, bird_type, beak):
        super().__init__(weight, colour, size, bird_type, beak)

    def fly(self):
        print("\nSparrow is flying")


class Penguin(Bird, NonFlyable):
    def __init__(self, weight, colour, size, bird_type, beak):
        super().__init__(weight, colour, size, bird_type, beak)


"""
As now we have solved Substitution problem 
"""

# LETS CALL EAGLE WITH STRONG BEAK CLASS
strong_beak = StrongBeak(20) # as it accepts strong beak 
eagle = Eagle(weight=10, colour=Color.BLUE, size=Size.LARGE, bird_type=BirdType.EAGLE, beak=strong_beak)

# LETS SAY TOMMAROW I WANT TO PASS WEAKBEAK TO EAGLE, HENCE WE WILL NOT BE PASSING REQUIRED BEAK TYPE 
weak_beak = WeakBeak(10)
eagle = Eagle(weight=10, colour=Color.BLUE, size=Size.LARGE, bird_type=BirdType.EAGLE, beak=weak_beak)

# SOLVING LISKOV SUBSTITUTION PROBLEM
# WE SHOULD HAVE ASSIGNED EAGLE WITH BEAK CLASS

````
